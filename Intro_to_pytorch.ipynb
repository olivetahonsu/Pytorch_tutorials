**Tensor:** A torch.Tensor is a multi-dimensional array of the same containing elements of the same data type.

- It similar to Numpy array but with more power.
- Can work both on CPU
and GPU.
- The defualt data type is float32.
- More suitable for deep learning than Numpy

```python
# Import torch
import torch
```

```python
# Create tensor
x = torch.tensor([1.0, 2.0, 3.0])
```

```python
print(x) # output tensor

print(x.shape) # shape of tensor

print(x.dtype) # data type of tensor

print(type(x)) # type of tensor
```

```python
# Create a 1D tensor with random values
y = torch.rand(3)

print(y)
```

- **Math Operations on tensors**

```python
a = torch.tensor([[1.0, 2.0, 3.0], [1.0, 3.0, 4.0]])
b = torch.tensor([4.0, 5.0, 6.0])

# Sum of tensors
print(a + b)

# Multiplication of tensors
print(a * b)

# Subtraction of tensors
print(a - b)

# Division of tensors
print(a/b)

# Mean of tensor
print(a.mean())
```

```python
# Tensor with power
x = torch.tensor([10.0, 20.0, 30.0])
y = x ** 2
print(y)
print(y.shape)
```

- **Automatic Differentiation (Autograd) in tensors**

```python

x = torch.tensor(2.0, requires_grad=True)
y = x ** 2

y.backward()

print(x.grad)
```

- **Reshape and View**

```python
my_tensor = torch.tensor(range(30))

my_tensor
```

```python
# Reshape tensor
reshaped_tensor = my_tensor.reshape(1, 30)

reshaped_tensor
```

```python
# Change a tensor value
reshaped_tensor[0, 0] = 100

reshaped_tensor
```

```python
# Grab an item in tensor
my_tensor[0]
```
